```python
import openpyxl
```

```python
wb = openpyxl.load_workbook('cBiT_config_v4_Excel_template.xlsx')
```

```python
ws = wb['studySample']
```

```python
fieldDefs = {}

for row in ws['A1:EC1']:
    for cell in row:
        fieldName = cell.value
        
        if fieldName.startswith("Characteristics["):
            fieldName = fieldName[len("Characteristics["):-1]
        elif fieldName.startswith("Factor Value["):
            fieldName = fieldName[len("Factor Value["):-1]
        
        fieldDefs[fieldName] = {
            'description': cell.comment.text,
            'data_type': 'string',
            'visibility': 'main',
            'category': 'Technical'
        }
        
import json
print(json.dumps(fieldDefs, indent=2, sort_keys=True))
```
